```python
import mediapipe as mp
import cv2
import numpy as np
import time
import random

def opponent(video, odd_even):
    mp_hands = mp.solutions.hands
    hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.7)
    
    required_hold_time = 3  # Hold your gesture for 3 seconds
    show_opponent_delay = 2  # Delay before showing opponent number

    tip_id = [4, 8, 12, 16, 20]
    current_count = None
    detection_start_time = None
    random_num = None
    show_opponent_time = None

    while True:
        ret, img = video.read()
        if not ret:
            continue

        frame = cv2.flip(img, 1)
        frame = cv2.resize(frame, None, fx=1.5, fy=1.5)
        img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        result = hands.process(img_rgb)
        lst = []
        finger_count = 0

        if result.multi_hand_landmarks:
            for hand in result.multi_hand_landmarks:
                for id, lm in enumerate(hand.landmark):
                    lst.append([id, lm.x, lm.y])

                if len(lst) == 21:
                    fingers = []

                    for i in range(1, 5):  # Index to pinky
                        if lst[tip_id[i]][2] < lst[tip_id[i] - 2][2]:
                            fingers.append(1)

                    # Thumb logic
                    if lst[4][0] < lst[3][0]:
                        fingers.append(1)

                    # Optional thumbs-up logic
                    thumbs_up = lst[4][2] < lst[3][2]
                    other_down = all(lst[tip_id[i]][2] > lst[tip_id[i] - 2][2] for i in range(1, 5))
                    if thumbs_up and other_down:
                        fingers = [1] * 6

                    finger_count = sum(fingers)

                    # Begin hold time check
                    if finger_count == current_count:
                        if detection_start_time is None:
                            detection_start_time = time.time()
                        elapsed = time.time() - detection_start_time
                        cv2.putText(frame, f"Holding: {int(elapsed)}s", (50, 450),
                                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

                        if elapsed >= required_hold_time:
                            # Generate opponent number once after hold
                            if random_num is None:
                                random_num = random.randint(0, 6)
                                show_opponent_time = time.time()  # start delay timer
                    else:
                        current_count = finger_count
                        detection_start_time = time.time()

        # If random_num is generated, wait 2 seconds then show it
        if random_num is not None:
            delay_elapsed = time.time() - show_opponent_time
            if delay_elapsed >= show_opponent_delay:
                cv2.putText(frame, f"Opponent: {random_num}", (30, 330),
                            cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 0, 0), 3)
                cv2.putText(frame, f"You: {finger_count}", (30, 400),
                            cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 0), 3)
                
                # End after showing both
                cv2.imshow("Hands", frame)
                cv2.waitKey(2000)  # show for 2 more seconds
                hands.close()
                cv2.destroyWindow("Hands")
                return finger_count, random_num

        else:
            # Ongoing message
            cv2.putText(frame, f"Hold gesture for {required_hold_time} seconds", (30, 90),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 255), 2)

        cv2.imshow("Hands", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    hands.close()
    cv2.destroyWindow("Hands")
    return None, None
```

```python
import cv2
import mediapipe as mp

# 👋 Load animation/video from local file
animation_path = "C:/Users/akhil/Downloads/Animation - 1751006648914.mp4"  # ← Change to your local video path
animation_cap = cv2.VideoCapture(animation_path)

# 🎯 Set up MediaPipe for hand detection
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(min_detection_confidence=0.7, max_num_hands=1)

# 🎥 Start webcam
cap = cv2.VideoCapture(0)

show_animation = False
animation_playing = False

while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame = cv2.flip(frame, 1)

    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(rgb)

    # Detect hand to trigger animation
    if result.multi_hand_landmarks:
        show_animation = True
        animation_cap.set(cv2.CAP_PROP_POS_FRAMES, 0)  # Restart video
        animation_playing = True

        for hand_landmarks in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

    # If animation is active
    if show_animation and animation_playing:
        anim_ret, anim_frame = animation_cap.read()
        if anim_ret:
            # Resize animation frame
            anim_frame = cv2.resize(anim_frame, (200, 200))

            # Overlay at top-left corner
            h, w, _ = anim_frame.shape
            frame[0:h, 0:w] = anim_frame
        else:
            # Animation ended
            show_animation = False
            animation_playing = False

    # Display final frame
    cv2.imshow("Gesture + Animation", frame)

    # Exit on ESC
    if cv2.waitKey(1) & 0xFF == 27:
        break

# Cleanup
cap.release()
animation_cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np
import time
import random

def detect_number_gestures(target_count=6):
    mp_hands = mp.solutions.hands
    hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.7)
    tip_ids = [4, 8, 12, 16, 20]
    values = []
    system_values = []

    cap = cv2.VideoCapture(0)
    storing = False
    black_screen_start = None
    last_user_gesture = None
    last_system_number = None

    while len(values) < target_count:
        ret, frame = cap.read()
        if not ret:
            continue

        frame = cv2.flip(frame, 1)
        h, w, _ = frame.shape
        img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        result = hands.process(img_rgb)

        # Black screen logic
        if storing:
            black_elapsed = time.time() - black_screen_start
            if black_elapsed < 1:
                black_frame = np.zeros_like(frame)
                cv2.putText(black_frame, f"Stored: {last_user_gesture}", (70, 220),
                            cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255, 255, 255), 3)
                cv2.putText(black_frame, f"System: {last_system_number}", (70, 300),
                            cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 255, 255), 3)
                cv2.imshow("Gesture Capture", black_frame)
                if cv2.waitKey(1) & 0xFF == ord('q'):
                    break
                continue
            else:
                storing = False  # Move to next round
                continue

        if result.multi_hand_landmarks:
            for handLms in result.multi_hand_landmarks:
                lm_list = [(lm.x, lm.y) for lm in handLms.landmark]
                if len(lm_list) == 21:
                    fingers = []

                    # Thumb
                    fingers.append(1 if lm_list[4][0] < lm_list[3][0] else 0)

                    # Four fingers
                    for i in range(1, 5):
                        fingers.append(1 if lm_list[tip_ids[i]][1] < lm_list[tip_ids[i] - 2][1] else 0)

                    user_count = sum(fingers)

                    if user_count != 0 and (len(values) == 0 or user_count != values[-1]):
                        values.append(user_count)
                        last_user_gesture = user_count
                        last_system_number = random.randint(1, 6)
                        system_values.append(last_system_number)
                        black_screen_start = time.time()
                        storing = True
                        print(f"✅ Stored {len(values)}: You = {user_count}, System = {last_system_number}")
                        break

                    cv2.putText(frame, f"Gesture: {user_count}", (50, 430),
                                cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255, 0, 0), 3)

        cv2.putText(frame, f"Show Gesture {len(values)+1}/{target_count}", (50, 100),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 255), 2)

        cv2.imshow("Gesture Capture", frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    hands.close()
    cv2.destroyAllWindows()

    print("\n✅ All Rounds:")
    for i in range(len(values)):
        print(f"  Round {i+1}: You = {values[i]}, System = {system_values[i]}")

    return values, system_values

# Run the gesture detector
detect_number_gestures()
```
